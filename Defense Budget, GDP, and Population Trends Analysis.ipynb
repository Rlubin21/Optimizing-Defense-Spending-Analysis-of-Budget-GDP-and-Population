```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('MilitarySpending.csv')
df.dropna(inplace=True)  
df = df
```

```python
# Converts columns to appropriate data types 
df['Year'] = df['Year'].astype(int)
df['DefenseBudget'] = df['DefenseBudget'].astype(float)
df['GDP'] = df['GDP'].astype(float)
df['Population'] = df['Population'].astype(float)
```

```python
df.head()
```

```python
# Plots the defense budget over time
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['DefenseBudget'], marker='o', linestyle='-', color='b')
plt.title('Defense Budget Over the Years')
plt.xlabel('Year')
plt.ylabel('Defense Budget (in billions)')
plt.grid(True)
plt.show()
```

```python
'''From 1960 to 2020, the defense budget showed a steady increase from $45 billion to approximately $770 billion.
This trend indicates a growing commitment to national defense and military readiness, reflecting broader" 
geopolitical tensions and strategic priorities.'''
```

```python
# Creates a new column for Defense Budget as a percentage of GDP
df['DefenseBudget_GDP_Percentage'] = (df['DefenseBudget'] / df['GDP']) * 100

# Plots Defense Budget as a percentage of GDP over the years
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['DefenseBudget_GDP_Percentage'], marker='o', linestyle='-', color='g')
plt.title('Defense Budget as Percentage of GDP Over the Years')
plt.xlabel('Year')
plt.ylabel('Defense Budget as % of GDP')
plt.grid(True)
plt.show()
```

```python
'''The economy is growing faster than defense spending, this may imply defense budget reductions or
economic expansion without parallel growth in military investments.'''
```

```python
# Creates a new column for Defense Budget per capita
df['DefenseBudget_PerCapita'] = df['DefenseBudget'] / df['Population']

# Plots Defense Budget per capita over time
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['DefenseBudget_PerCapita'], marker='o', linestyle='-', color='r')
plt.title('Defense Budget Per Capita Over the Years')
plt.xlabel('Year')
plt.ylabel('Defense Budget Per Capita (in $)')
plt.grid(True)
plt.show()
```

```python
'''This indicates that the government is allocating more defense resources per person.
 This could reflect a growing focus on military investment, possibly due to geopolitical threats,
 military modernization, or ongoing conflicts.'''
```

```python
# Sorts the data by Efficiency (Defense Budget as % of GDP)
df_sorted = df.sort_values(by='DefenseBudget_GDP_Percentage', ascending=False)

# Displays the top 5 years with the highest defense budget as a % of GDP
print(df_sorted[['Year', 'DefenseBudget_GDP_Percentage']].head())
```

```python
# Sorts data by Defense Budget as a percentage of GDP in ascending order (lowest first)
df_sorted_lowest_efficiency = df.sort_values(by='DefenseBudget_GDP_Percentage', ascending=True)

# Displays the top 5 years with the lowest defense budget as a percentage of GDP
print(df_sorted_lowest_efficiency[['Year', 'DefenseBudget_GDP_Percentage']].head())
```

```python
'''1999, 2000 and 2001 were periods of high efficiency in terms of defense spending relative to the size of the economy. 
 These years saw the defense budget consuming a smaller portion of GDP, 
 which could suggest a more efficient allocation of resources.
 This is likely due to strong economic performance or strategic shifts.'''
```

```python
# Sorts the data by Defense Budget Per Capita
df_sorted_percapita = df.sort_values(by='DefenseBudget_PerCapita', ascending=False)

# Displays the top 5 years with the highest defense budget per capita
print(df_sorted_percapita[['Year', 'DefenseBudget_PerCapita']].head())
```

```python
'''2011, 2010, 2020 were periods of low efficiency often associated with military conflicts, geopolitical tensions, 
 or stagnating economic growth.'''
```

```python
# Correlation matrix
corr_matrix = df[['DefenseBudget', 'GDP', 'Population']].corr()

# Plots a heatmap for the correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Defense Budget, GDP, and Population')
plt.show()
```
